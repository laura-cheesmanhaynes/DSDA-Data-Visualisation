# **Plotting with Seaborn**

Documentation of Seaborn: https://seaborn.pydata.org/

For choosing a list of colors or continuous colormap defining a palette refer to this link: https://seaborn.pydata.org/generated/seaborn.color_palette.html

For choosing colour hues with constant lightness and saturation in the HUSL system refer to this link: https://seaborn.pydata.org/generated/seaborn.husl_palette.html

```python
# Import Seaborn librabry
import seaborn as sns
```

```python
# Get names of dataset that come with Seaborn library
sns.get_dataset_names()
```

```python
# Exploring individual dataset -->

# Load data
tips = sns.load_dataset("tips")

# Display the first few rows
print("\n\n### Display the first few rows --> \n")
print(tips.head())

# Display basic information about the DataFrame
print("\n\n### Display basic information about the DataFrame --> \n")
print(tips.info())

# Display descriptive statistics
print("\n\n### Display descriptive statistics --> \n")
print(tips.describe())

# Display the count of unique values in each column
print("\n\n### Display the count of unique values in each column --> \n")
print(tips.nunique())
```

```python
# Relational plot -->

# Load data
tips = sns.load_dataset("tips")

# Relational plot
sns.relplot(x="total_bill", y="tip", data=tips)

# Scatter plot with a custom color palette
sns.relplot(data=tips, x="total_bill", y="tip", hue="time", palette="husl")
```

```python
# Scatter plot -->

# Load data
iris = sns.load_dataset("iris")

# Scatter plot
sns.scatterplot(data=iris, x="sepal_length", y="petal_length", hue="species")

# Scatter plot with a specific color for markers
# sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex", palette="Set1")
```

```python
# Line plot -->

# Load data
flights = sns.load_dataset("flights")

# Line plot
sns.lineplot(data=flights, x="year", y="passengers")
```

```python
# Histogram using displot() -->

# Load data
tips = sns.load_dataset("tips")

# Histogram using displot()
sns.displot(tips["total_bill"], kde=True)
```

```python
# Histogram using histplot() -->

# Load data
tips = sns.load_dataset("tips")

# Histogram using histplot()
sns.histplot(tips["total_bill"], kde=True)
```

```python
# Plotting Bivariate Data with Jointplot() -->

# Load data
iris = sns.load_dataset("iris")

# Joint plot
sns.jointplot(data=iris, x="sepal_length", y="petal_length", kind="scatter")
```

```python
# Plotting Many Distribution Data with Pairplot() -->

# Load data
iris = sns.load_dataset("iris")

# Pairplot
sns.pairplot(iris)
```

```python
# Plotting Distributions with Rugplots() -->

# Load data
tips = sns.load_dataset("tips")

# Rugplot
sns.rugplot(data=tips, x="total_bill", height=0.7)  # Adjust the height parameter as needed
```

```python
# Creating interactive plot with Plotly -->
import plotly.express as px
# Documentation of Plotly can be accessed here: https://plotly.com/python/

# Load data
tips = sns.load_dataset("tips")

# Create an interactive scatter plot using Plotly
fig = px.scatter(tips, x="total_bill", y="tip", color="sex", hover_data=["day", "time", "size"])

# Show the plot
fig.show()
```
