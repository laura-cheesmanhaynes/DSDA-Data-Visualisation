```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the 'salessheet' into a DataFrame
sales_df = pd.read_excel('/content/gdrive/My Drive/YSJ-DataVis/DimGeography.xlsx', sheet_name='salessheet')

# Line Plot
plt.figure(figsize=(10, 6))
plt.plot(sales_df['StoreID'], sales_df['Sales'], marker='o', linestyle='-', color='b')
plt.title('Line Plot of Store ID vs. Sales')
plt.xlabel('Store ID')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

```python
# Scatter Plot
plt.figure(figsize=(10, 6))
plt.scatter(sales_df['StoreID'], sales_df['Sales'], color='r')
plt.title('Scatter Plot of Store ID vs. Sales')
plt.xlabel('Store ID')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

```python
# Bar Plot
plt.figure(figsize=(10, 6))
sales_df['State'].value_counts().plot(kind='bar', color='g')
plt.title('Bar Plot of State')
plt.xlabel('State')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
# Histogram of Sales
plt.figure(figsize=(10, 6))
plt.hist(sales_df['Sales'], bins=20, color='m', edgecolor='k')
plt.title('Histogram of Sales')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Pie Chart
plt.figure(figsize=(10, 6))
sales_by_state = sales_df['State'].value_counts()
plt.pie(sales_by_state, labels=sales_by_state.index, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart of Sales by State')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Box Plot
plt.figure(figsize=(10, 6))
plt.boxplot(sales_df['Sales'], vert=False)
plt.title('Box Plot of Sales')
plt.xlabel('Sales')
plt.grid(True)
plt.show()
```

**Image Plotting**

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a random 2D array representing an image
image_data = np.random.rand(100, 100)

# Plot the image
plt.figure(figsize=(8, 6))
plt.imshow(image_data, cmap='gray')  # 'gray' colormap for grayscale image
plt.title('Random Image')
plt.colorbar()  # Add colorbar to show intensity scale
plt.axis('off')  # Turn off axis
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create a 3D array representing a colorful image
image_data = np.random.randint(0, 256, size=(100, 100, 3), dtype=np.uint8)

# Plot the image
plt.figure(figsize=(8, 6))
plt.imshow(image_data)
plt.title('Colorful Image')
plt.axis('off')  # Turn off axis
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
sales_df = pd.read_excel('/content/gdrive/My Drive/YSJ-DataVis/DimGeography.xlsx', sheet_name='salessheet')

# Filter sales data for StoreID 1 to 7
sales_data = sales_df.loc[sales_df['StoreID'].isin([1, 2, 3, 4, 5, 6, 7])]

# Pivot the data to get StoreID as columns and Years as rows
pivot_sales = sales_data.pivot_table(index='Year', columns='StoreID', values='Sales', aggfunc='sum')
print(pivot_sales)

# Create the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_sales, cmap='viridis', annot=True, fmt='.0f')
plt.title('Sales Heatmap by StoreID and Year')
plt.xlabel('StoreID')
plt.ylabel('Year')
plt.show()
```
